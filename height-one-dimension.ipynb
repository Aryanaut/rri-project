```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Constants and inputs
g = 9.8

u = int(input("Initial Velocity: "))
```

```python
# Setting up graph
fig, ax = plt.subplots()
plt.title("h(t)")
plt.xlabel("Time(t) in seconds")
plt.ylabel("Height(h) in meters")

xCoords = [0]
yCoords = [0]
h = 0
t0 = 0
maxH = (u ** 2) / (2 * g)

while h >= 0:
    xCoords.append(t0)
    h = 0 + (u * t0) - (0.5 * g * (t0 * t0))
    yCoords.append(h)
    t0 += 0.1
    
ax.plot(xCoords, yCoords)
plt.show()
```

```python
'''The function h(t) can be given by h(t) = u*t - 0.5*g*(t^2)
    g = 9.8 m/s^2 and u is the initial velocity'''
```
